### Null vs Alternative Hypothesis

There are 4 steps in Data-Driven Decision Making
 - Formulate a hypothesis
 - Find the right test for the hypothesis
 - Execute the test 
 - Make a decision based on the result

#### Hypothesis:
It is an idea that can be tested. When there is no data to be tested, it is not a hypotesis.

Types of Hypothesis:

Null Hypothesis
 - Denoted as H0 => 
 - the one to be tested
 - The concept is "innocent until proven guilty"
 - it is the statement we try to reject
 - it is the status quo or the present state of affairs.
 
Alternative Hypothesis:
 - Denoted as H1 or HA.
 - it everything else the null hyp is not
 - it is the change or innovation that challenges the status quo
 - it is our personal opinion
 
Example 1:

Null Hyp: Mean data scientist salary equals 113,000 dollars

Alternative Hyp = Mean data scientist salary is not equal to 113,000 dollars

To check if the 113,000 is close enough to the true mean predicted by our sample

Accept if: Sample is close enough to the true mean

Reject if: Sample is too far from the true mean

This is an example of a two-sided test

Example 2:

The mean data scientists salary is more than 125,000 dollars

Null Hyp: Mean data scientist salary is more or equal to 125,000

Alternative Hyp: Mean data scientist salary is less than 125,000

Outcomes of tests refer to population parameter rather than the sample statistic. The result we get is for the population.

### Rejection region and significance level

Significance level:

 - It is the probability of rejecting the null hypothesis if it is true.
 - Denoted by Alpha. 
 - Typical levels for alpha are: 0.01, 0.05, 0.1 - a value we select based on the certainty we need
 - The choice of alpha is determined by the context we are operating in. 0.05 is the most commonly used value.
 
Example 1: To test if a machine works properly. We expect the machine to make little or no mistakes. To be very precise, pick a low significance level such as 0.01.

Example 2: Carrying out an analysis on students performance on the average as a researcher.
The Dean believes on average, students GPA is 70%. because you cant agree with his opinion, you start testing.

Null Hyp (H0): Population mean grade is 70%

Alt. Hyp (H1): Population mean grade is not 70% 

The test to perform is the Z-test = Sample mean - Hypothesized mean/standard error.

 - if the sample mean is close to the hypothesized mean, Z will be close to zero. otherwise, it will be far away from it.
 - But if the sample mean = hypothesized mean, Z = 0. Accept null hyp. in both cases.
 
How big should Z be to reject the null hyp?

Conducting a 2-sided/tail test, there are 2 cut off lines,one on each sides. If Z falls in the middle part of the normal distribution, accept the null hyp. If it falls outside, reject the null hyp.
The shaded part (the part outside) is called the REJECTION REGION which depends on the significance level.
If alpha = 0.05, alpha/2 = 0.025. Z = 1.96. Meaning Z is between -1.96 and 1.96.
Z-Test < -1.96 or > 1.96 => Reject the Null Hyp. Otherwise, accept it.

How Hypothesis Testing works:

- Calculate a statistic (e.g, sample mean)
- scale it e.g using the z-test
- check if Z is in the rejection region (close to 0, accept null. Far from 0, reject it.)

One-sided Test

Example 1: A statement that data scietists make more than 125,000 dollars.

Null: >= 125,000

Alt: < 125,000

significance level = 0.05.

Z = -1.645 (left rejection region). If Z < -1.645, Reject the Null. Otherwise, accept.

Example 2: A statement that the average students GPA is lower than 70%

Null: <= 70%

Alt: > 70%

significance level = 0.05.

Z = 1.645 (right rejection region). If Z > 1.645, Reject the Null. Otherwise, accept.

### Type I error vs type II error (errors that cna be made in Hypothesis testing)

Type I error: 

 - This is made when you reject the true null hypothesis
 - Also called false positive
 - Probability of making this error is Alpha the level of significance.
  
  
Type II error

 - This is made when you accept the false null hypothesis
 - Also called false positive
 - Probability of making this error is denoted by beta the level of significance.
 - Beta depends mainly on sample size and magnitude of the effect.
 - Probabilty of rejecting a false null hypothesis (Probability = 1 - beta) a.k.a. power of the test. Researchers increase the power of the test by increasing sample size.

### P-VALUE

This is the most common way to test hypothesis. Instead of testing at pre-assigned level of significance. P-value is the smallest level of significance at which we can still reject the null hypothesis, given the oobserved sample statistic.

Example: Using the data scientist salary.

Null: Pop. mean (before) >/= Pop. mean (after) 

Normal distribution

Z = -4.67

We rejected the null at 0.05 and 0.01. How much lower can we go?

P-value = 0.0001

Rule: Reject the null if P-value < alpha (significance level).

 - Test at 90%: 0.0001 < 0.1

 - Test at 95%: 0.0001 < 0.005

 - Test at 99%: 0.0001 < 0.01

Therefore, reject null hypothesis.

2. Z = 2.12

Rule: Reject the null if P-value < alpha

 - Test at 90%: Reject

 - Test at 95%: Reject

 - Test at 99%: Cannot reject

Look for the p-value on the table

To find p-value manually:

1. One-sided test: p-value is 1 - the number from the table. 1 - 0.983 = 0.017.

2. Two-sided test: p-value is 1 - the number from the table multiplied by 2. (1 - 0.983) * 2 = 0.034.

Where and How are P-Values used:

1. Most statistical software calculates p-values for each test.

2. Researcher decides significance post-factum.

3. P-values are usually found with 3 digits after the dot x.xxx

4. The closer the 0.000 the p-value, the better the result.

P-value is a universal concept that works with every distribution:

 - Normal
 - Binomial
 - Student's T
 - Uniform

If the P-value is lower than the level of significance - reject the null hypothesis.

Use online p-value calculators to support your studies.

### Test for the mean. Population variance unknown

Example: As a marketing Analyst, estimate if our competitor has a higher open rate. (Note: E-mail open rate measures gow many people on the email list actually opened an email.

Where your company's open rate = 40%

Hypothesis: 
Null: mean open rate </= 40%
Alt: Mean open rate > 40% (a one-sided test)

In hypothesis testing, we aim to reject the null hypothesis.

@ 0.05 level of significance

Test where:

 - Sample mean = 37.70%
 - Sample std = 13.74%
 - Standard error = 4.34%

Use T-statistic because the variance is small and the sample is not big enough.

T = 37.70% - 40% / 4.34% = -0.53
Compare absolute value of -0.53 with
degrees of freedom = n - 1 = 9
0.05 one-sided significance = 0.53

T = 0.53 < t = 1.83. 
Therefore, accept the null hypothesis. This means that, at this level of significance statistically, we cannot say that the e-mail open rate is higher than 40%.

Note: Accept if: T-score (T) < critical score (t)
      Reject if: T-score (T) > critical score (t)
      
P-value = 0.304 > 0.05 => Accept the null

 - Accept if: p-value > alpha (significance level)
 - Reject if: p-value < alpha (significance level)

P-value = 0.304 > 0.01 => Accept the null

### Test for the mean. Dependent samples

For Multiple populations - Dependent samples. E.g weight loss or blood tests.

Example: using the example of magnesium levels in one's blood with 10 people. We will conclude on the effectiveness of the drug (is it working?)

Soultion:
Null: Pop. mean (before) >/= Pop. mean (after) 
Alt: Pop. mean (before) </= Pop. mean (after)

To know if the magnesium levels are higher:
 - Sample mean = -0.33%
 - Sample std = 0.45%
 - Standard error = 0.14%

Using the t-statistic
 - small sample
 - assuming that it is a normal distribution 
 - variance unknown
T = -0.33 - 0 / 0.14 = -2.29
p-value = 0.024

Decision rule: Accept if p-value > alpha
               Reject if p-value < alpha

@ 5% significance = 0.024 < 0.05 => Reject null
@ 1% significance = 0.024 > 0.01 => Accept null 

The lowest significance level at which we can reject the null is 0.024 which is exactly the p-value
